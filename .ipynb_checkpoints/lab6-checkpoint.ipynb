### Подключим необходимые библиотеки и настроим датасет MNIST :

```python
import torch
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

train_dataset = datasets.MNIST(root='./data', train=True, transform=transforms.ToTensor(), download=True)
test_dataset = datasets.MNIST(root='./data', train=False, transform=transforms.ToTensor(), download=True)
```

### Преобразуем набор данных в массивы numpy для построения дерева решений

```python
train_data = train_dataset.data.numpy().reshape(-1, 28*28)
train_labels = train_dataset.targets.numpy()
test_data = test_dataset.data.numpy().reshape(-1, 28*28)
test_labels = test_dataset.targets.numpy()
```

### Создадим объект классификатора дерева решений и обучает классификатор дерева решений на train:

```python
tree = DecisionTreeClassifier(max_depth=10, min_samples_split=10, min_samples_leaf=2, random_state=42)
tree.fit(train_data, train_labels)
```

### Выводим точность классификатора дерева решений на тестовых данных:

```python
test_accuracy = tree.score(test_data, test_labels)
print(f'Test accuracy: {test_accuracy*100:.2f} %')
```

### Визуализируем дерево принятия решений с помощью функции из библиотеки scikit-learn:

```python
plot_tree(tree)
plt.show()
```
